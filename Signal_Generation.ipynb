```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def new_signal(freq,time):
  sin_1 = np.sin(2*np.pi*(freq*time))
  cos_2 = np.cos(2*np.pi*(2*freq*time))
  sin_3 = np.sin(2*np.pi*(3*freq*time))
  return sin_1 + cos_2 + sin_3
```

```python
def plot_signal(signal,time):
  plt.figure(figsize = (20,7))
  plt.plot(signal,time)
  plt.xlabel("time")
  plt.ylabel("Amplitude")
  plt.show()
```

```python
time = np.linspace(0,10,1000)
signal = new_signal(freq=1,time=time)
plot_signal(time,signal)
```
